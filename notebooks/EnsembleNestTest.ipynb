## A Simple 1D Test

Now we define some priors and likelihoods.  First, the prior, which we take to be uniform in $[0,1]$.  However, for sampling efficiency, we transform $[0,1]$ to $[-\infty, \infty]$ using the logit transform.  If we write $y = \mathrm{logit}(x)$, with $p(x) = 1$, then 
$$
p(y) = \frac{p(x)}{dy/dx} = \frac{1}{\frac{1}{x} + \frac{1}{1-x}} = x(1-x) = \frac{e^y}{\left(1 + e^y\right)^2}
$$

```julia
function logp(y)
    if y < zero(y)
        y - 2.0*log1p(exp(y))
    else
        -y - 2.0*log1p(exp(-y))
    end
end

function logp(y::Array{Float64,1})
    logp(y[1])
end
```

```julia
ys = collect(-10:0.1:10)
plot(x=ys, y=[logp(y) for y in ys], Geom.line)
```

And the likelihood is N(0.5, 0.01):

```julia
function logl(y)
    sigma = 0.01
    mu = 0.5
    
    ey = exp(y)
    x = ey / (one(ey) + ey)
    
    -0.5*log(2.0*pi) - log(sigma) - 0.5*(x-mu)*(x-mu)/(sigma*sigma)
end

function logl(y::Array{Float64, 1})
    logl(y[1])
end
```

```julia
plot(x=exp(ys)./(1.0+exp(ys)), y=[logl(y) for y in ys], Geom.line)
```

Now we set up and run nested sampling:

```julia
xs = rand((1, 512))
ys = log(xs) - log(1-xs)
ns = Ensemble.EnsembleNest.NestState(logl, logp, ys, 100)
```

```julia
Ensemble.EnsembleNest.run!(ns, 1e-3)
```

```julia
post = Ensemble.EnsembleNest.postsample(ns)
println(size(post,2))
post = reshape(post, (size(post, 2),))
post = exp(post)./(1.0 + exp(post)) # Convert back to [0,1]
```

```julia
xs = collect(0.45:0.001:0.54)
plot(layer(x=post, Geom.histogram(density=true, )), layer(x=xs, y=1.0/sqrt(2*pi)/0.01*exp(-0.5*(xs-0.5).*(xs-0.5)/0.01/0.01), Geom.line))
```

```julia
Ensemble.EnsembleNest.logZ(ns), Ensemble.EnsembleNest.logdZ(ns)
```

```julia
plot(x=1:size(ns.deadpts, 2), y=ns.deadlogls, Geom.line)
```

## Brewer Likelihood

OK.  That was pretty easy.  Let's try the Brewer likelihood, from [Brewer, et al (2011)](http://arxiv.org/abs/0912.2380), which is a lot harder to sample.  We start by defining the distribution that Brewer used.  The ``make_logpl()`` function uses a trick that ensures that the "global" variables are actually local and therefore enabling a lot more optimisation

```julia
function make_logpl()
    ndim = 20
    v = 0.1
    u = 0.01
    x0 = 0.031
    pwidth = 1
    
    function logp_brewer(x)
        if any(x .< -pwidth/2.0) || any(x .> pwidth/2.0)
            -Inf
        else
            0.0
        end
    end
    
    function logl_brewer(x)        
        g1 = -0.5*ndim*log(2.0*pi) - ndim*log(v) - 0.5*sum(x.*x)/(v*v)
        g2 = -0.5*ndim*log(2.0*pi) - ndim*log(u) - 0.5*sum((x - x0).*(x - x0))/(u*u)
    
        logsumexp(g1, log(100.0)+g2)
    end    
    
    logl_brewer, logp_brewer
end
logl_brewer, logp_brewer = make_logpl()
```

```julia
ns = NestState(logl_brewer, logp_brewer, -0.5+rand((20, 1000)), 100)
```

```julia
run!(ns, 1e-12)
```

```julia
plot(x=-collect(1:size(ns.deadlogls, 1))/float(size(ns.livepts, 2)), y=ns.deadlogls, Geom.line)
```
